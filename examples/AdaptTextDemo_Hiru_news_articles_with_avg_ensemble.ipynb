```python
from AdaptTextLib.adapttext.adapt_text import AdaptText

lang = 'si'
app_root = "/storage"
bs = 128
splitting_ratio = 0.1
adapttext = AdaptText(lang, app_root, bs, splitting_ratio)
```

```python
adapttext.prepare_pretrained_lm("full_si_dedup.zip")
```

```python
import pandas as pd

pd.set_option('display.max_colwidth', -1)
# path_to_csv="sinhala-hate-speech-dataset.csv"
path_to_csv="Sinhala_news_articles.csv"
# path_to_csv="small_sin.csv"
df = pd.read_csv(path_to_csv)

text_name = "Title"
label_name = "Label"

# print(df.head())

classifierModelFWD, classifierModelBWD, classes = adapttext.build_classifier(df, text_name, label_name, grad_unfreeze=True)
```

```python
from AdaptTextLib.adapttext.evaluator.evaluator import Evaluator

evaluator = Evaluator()
evaluator.evaluate_ensemble(classifierModelFWD, classifierModelFWD)
```
